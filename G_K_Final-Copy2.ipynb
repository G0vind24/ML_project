```python
x = df.iloc[:, df.columns != 'Action']
x.head()
```

```python
y = df.iloc[:, df.columns == 'Action']
y.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.20, random_state=57)
```

```python
from sklearn.preprocessing import StandardScaler

# Initializing the StandardScaler
scaler = StandardScaler()

# Fitting the scaler on the training data and transforming the training data
x_train = scaler.fit_transform(x_train)

# Transforming the test data using the fitted scaler
x_test = scaler.transform(x_test)

# Displaying the transformed test data
x_test
```

```python
#logistic regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

# Initializing the Logistic Regression classifier
classifier = LogisticRegression(n_jobs=-1)

# Fitting the classifier to the training data
classifier.fit(x_train, y_train)

# Making predictions on the test data
y_pred = classifier.predict(x_test)

# Calculating the accuracy score for the model's predictions
lraccuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {lraccuracy}")

# Calculating precision, recall, and F1-score with weighted averaging
LRprecision_weighted = precision_score(y_test, y_pred, average='weighted')
LRrecall_weighted = recall_score(y_test, y_pred, average='weighted')
LRf1_weighted = f1_score(y_test, y_pred, average='weighted')
print(f"Precision : {LRprecision_weighted}")
print(f"Recall : {LRrecall_weighted}")
print(f"F1-score : {LRf1_weighted}")
```

```python

# Print the actual values of the top 5 rows from the test data
print("Actual values of the top 5 rows from test data:")
print(y_test[:5])

# Select the top five rows from x_test for prediction
LRP = x_test[:5]

# Make predictions using the logistic regression model
LPrediction = classifier.predict(LRP)

# Print the predictions for the top five rows
print("Predictions for the top 5 rows:")
print(LPrediction)
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Initialize and fit the Decision Tree classifier with a specified depth
treemodel = DecisionTreeClassifier(max_depth=3)
treemodel.fit(x_train, y_train)

# Plotting the Decision Tree
plt.figure(figsize=(15, 10))
tree.plot_tree(treemodel, filled=True)
plt.title('Decision Tree Visualization with Max Depth 3')
plt.show()

# Making predictions on the test data
y_pred = treemodel.predict(x_test)

# Calculating the accuracy score for the model's predictions
dtaccuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {dtaccuracy}")

# Calculating precision, recall, and F1-score with weighted averaging
dtprecision_weighted = precision_score(y_test, y_pred, average='weighted')
dtrecall_weighted = recall_score(y_test, y_pred, average='weighted')
dtf1_weighted = f1_score(y_test, y_pred, average='weighted')

print(f"Precision (weighted): {dtprecision_weighted}")
print(f"Recall (weighted): {dtrecall_weighted}")
print(f"F1-score (weighted): {dtf1_weighted}")
```

```python
# Print the actual values of the top 5 rows from the test data
print("Actual values of the top 5 rows from test data:")
print(y_test[:5])

# Select the top five rows from x_test for prediction
DTP = x_test[:5]

DTPrediction = treemodel.predict(DTP)

# Print the predictions for the top five rows
print("Predictions for the top 5 rows:")
print(DTPrediction)
```

```python

#Random Forest
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest Classifier
fit_rf = RandomForestClassifier(random_state=40)

# Train the model
fit_rf.fit(x_train, y_train)

# Make Predictions
predictions_rf = fit_rf.predict(x_test)

# Calculate performance metrics
rfaccuracy = accuracy_score(y_test, predictions_rf)
rfprecision_weighted = precision_score(y_test, predictions_rf, average='weighted')
rfrecall_weighted = recall_score(y_test, predictions_rf, average='weighted')
rff1_weighted = f1_score(y_test, predictions_rf, average='weighted')

# Print the performance metrics for Random Forest
print(f"Accuracy: {rfaccuracy}")
print(f"Precision (weighted): {rfprecision_weighted}")
print(f"Recall (weighted): {rfrecall_weighted}")
print(f"F1-score (weighted): {rff1_weighted}")
```

```python

# Print the actual values of the top 5 rows from the test data
print("Actual values of the top 5 rows from test data:")
print(y_test[:5])

# Select the top five rows from x_test for prediction
RFP = x_test[:5]

RFPrediction = fit_rf.predict(RFP)

# Print the predictions for the top five rows
print("Predictions for the top 5 rows:")
print(RFPrediction)
```

```python
#GaussianNB

# Import necessary libraries
from sklearn.naive_bayes import GaussianNB

# Initialize the Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Train the model
gnb.fit(x_train, y_train)

# Make Predictions
y_pred = gnb.predict(x_test)

# Calculate performance metrics
accuracyG = accuracy_score(y_test, y_pred)
precision_weighted_gnb = precision_score(y_test, y_pred, average='weighted')
recall_weighted_gnb = recall_score(y_test, y_pred, average='weighted')
f1_weighted_gnb = f1_score(y_test, y_pred, average='weighted')

# Print the performance metrics for Naive Bayes
print(f"Accuracy (Naive Bayes): {accuracyG}")
print(f"Precision (weighted): {precision_weighted_gnb}")
print(f"Recall (weighted): {recall_weighted_gnb}")
print(f"F1-score (weighted): {f1_weighted_gnb}")
```

```python

# Print the actual values of the top 5 rows from the test data
print("Actual values of the top 5 rows from test data:")
print(y_test[:5])

# Select the top five rows from x_test for prediction
GNP = x_test[:5]

GNPrediction = gnb.predict(GNP)


# Print the predictions for the top five rows
print("Predictions for the top 5 rows:")
print(GNPrediction)
```

```python

#KNN

from sklearn.neighbors import KNeighborsClassifier

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(x_train, y_train)

# Make Predictions
y_prediction = knn.predict(x_test)

# Calculate performance metrics
accuracy_knn = accuracy_score(y_test, y_prediction)
precision_weighted_knn = precision_score(y_test, y_prediction, average='weighted')
recall_weighted_knn = recall_score(y_test, y_prediction, average='weighted')
f1_weighted_knn = f1_score(y_test, y_prediction, average='weighted')

# Print the performance metrics for KNN
print(f"Accuracy (KNN): {accuracy_knn}")
print(f"Precision (weighted): {precision_weighted_knn}")
print(f"Recall (weighted): {recall_weighted_knn}")
print(f"F1-score (weighted): {f1_weighted_knn}")
```

```python
# Print the actual values of the top 5 rows from the test data
print("Actual values of the top 5 rows from test data:")
print(y_test[:5])

# Select the top five rows from x_test for prediction

KNNP = x_test[:5]

KNNPrediction = knn.predict(KNNP)


# Print the predictions for the top five rows
print("Predictions for the top 5 rows:")
print(KNNPrediction)
```

```python
#Support Vector Classifier

from sklearn.svm import SVC
# Initialize the Support Vector Classifier with a linear kernel
clf = SVC(kernel='linear')

# Train the model
clf.fit(x_train, y_train)

# Make predictions
y_pred = clf.predict(x_test)

# Calculate performance metrics
svc_accuracy = accuracy_score(y_test, y_pred)
precision_weighted_svc = precision_score(y_test, y_pred, average='weighted')
recall_weighted_svc = recall_score(y_test, y_pred, average='weighted')
f1_weighted_svc = f1_score(y_test, y_pred, average='weighted')

# Print the performance metrics for SVC
print("Accuracy (SVC):",svc_accuracy)
print("Precision (weighted):",precision_weighted_svc)
print("Recall (weighted):",recall_weighted_svc)
print("F1-score (weighted):",f1_weighted_svc)
```
